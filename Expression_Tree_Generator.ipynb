```python
# *** Expression Tree Generator and Evaluator ***

# Problem Statement:
# To construct an expression tree from an arithmetic expression
# and evaluate it using recursive traversal techniques.

# Data Structures Used:
# - Stack (for building tree)
# - Binary Tree
# - Recursion

# Algorithm Steps:
# 1. Take postfix expression as input.
# 2. Read tokens one by one.
# 3. If operand -> push to stack.
# 4. If operator -> pop two operands and create tree node.
# 5. Repeat until tree is formed.
# 6. Recursively evaluate tree.
```

```python
# Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class ExpressionTree:
    def __init__(self, expression):
        self.root = self.build_tree(expression)

    # Build tree from postfix expression
    def build_tree(self, expression):
        stack = []
        operators = ['+', '-', '*', '/']
        tokens = expression.split()

        for token in tokens:
            if token not in operators:
                stack.append(Node(token))
            else:
                node = Node(token)
                node.right = stack.pop()
                node.left = stack.pop()
                stack.append(node)

        return stack[0]

    # Inorder traversal
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.value, end=' ')
            self.inorder(node.right)

    # Recursive evaluation
    def evaluate(self, node):
        if node.left is None and node.right is None:
            return int(node.value)

        left_value = self.evaluate(node.left)
        right_value = self.evaluate(node.right)

        if node.value == '+':
            return left_value + right_value
        elif node.value == '-':
            return left_value - right_value
        elif node.value == '*':
            return left_value * right_value
        elif node.value == '/':
            return left_value / right_value


# Main Program
expression = input("Enter postfix expression (with spaces): ")

tree = ExpressionTree(expression)

print("\nInorder Traversal:")
tree.inorder(tree.root)

print("\nResult:", tree.evaluate(tree.root))
```
